```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
# tokeninzation -- paragraph into sentences
nltk.download('punkt')
```

```python
sentences =nltk.sent_tokenize(paragraph)  # split the whole paragraph into sentences
sentences
```

```python
stemmer = PorterStemmer()
```

```python
nltk.download('wordnet')
```

```python
# for lemmatization

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
print(f"stemmer - {stemmer.stem('history')} lemmatizer -- {lemmatizer.lemmatize('history')}")
```

```python
import re
corpus = []
for i in range(len(sentences)):
  text = re.sub('[^a-zA-Z0-9]',' ',sentences[i])
  corpus.append(text.lower())

corpus
```

```python
nltk.download('stopwords')
```

```python
# stemming
corp_without_sw=[]
for i in corpus:
  words = nltk.word_tokenize(i) #splits each i into individual words
  for word in words:
     if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

```python
#lemmatize

for i in corpus:
  words = nltk.word_tokenize(i)
  for word in words:
     if word not in set(stopwords.words('english')):
      print(lemmatizer.lemmatize(word))
```

```python
import re
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True,max_features=4)
```

```python
x =cv.fit_transform(corpus)
```
